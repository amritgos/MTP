```python
import pandas as pd
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
# matplotlib.use('Agg')
import datetime

%matplotlib inline
from finrl.apps import config
from finrl.neo_finrl.preprocessor.yahoodownloader import YahooDownloader
from finrl.neo_finrl.preprocessor.preprocessors import FeatureEngineer
import warnings
warnings.filterwarnings("ignore")
```

```python
df = YahooDownloader(start_date = '2016-01-01',
                     end_date = '2021-01-01',
                     ticker_list = config.SENSEX_30_TICKER).fetch_data()
```

```python
sentiment_df = pd.read_csv('datasets/Sentiment_Compiled.csv')
```

```python
fe = FeatureEngineer(
                    use_technical_indicator=False,
                    use_turbulence=False,
                    use_vix=False,
                    use_sentiment=True,
                    sentiment_df=sentiment_df,
                    user_defined_feature = True)

processed = fe.preprocess_data(df)
```

```python
processed
```

```python

import pandas as pd
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.stats as stats

for asset in df['tic'].unique():
    print(asset)
    df_asset = processed.loc[processed.tic == asset]

    column = 'news'
    df_asset[column] = (df_asset[column] - df_asset[column].mean()) / (df_asset[column].std())    
    column2 = 'close'
    df_asset[column2] = (df_asset[column2] - df_asset[column2].mean()) / (df_asset[column2].std())    
    
    r, p = stats.spearmanr(df_asset.dropna()[column2], df_asset.dropna()[column])
    print(f"Scipy computed Spearman r: {r} and p-value: {p}")

    # Compute rolling window synchrony
    f,ax=plt.subplots(figsize=(7,3))
    df_asset[[column,column2]].rolling(window=300,center=True).median().plot(ax=ax)
    ax.set(xlabel='Time',ylabel='Spearman r')
    plt.show()
    # ax.set(title=f"Overall Pearson r = {np.round(overall_pearson_r,2)}");
```
